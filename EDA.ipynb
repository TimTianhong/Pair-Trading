```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download historical data for JPM and GS
tickers = ['JPM', 'GS']
data = yf.download(tickers, start="2012-01-01", end="2017-12-31")['Adj Close']
```

```python
data['GS_normalized'] = (data['GS'] - data['GS'].mean()) / data['GS'].std()
data['JPM_normalized'] = (data['JPM'] - data['JPM'].mean()) / data['JPM'].std()
```

```python
data['Spread'] = data['GS_normalized'] - data['JPM_normalized']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['JPM_normalized'], label='JPM')
plt.plot(data['GS_normalized'], label='GS')
plt.title('JPM and GS Stock Prices (Normalized)')
plt.legend()
plt.grid(True)
plt.show()
```
